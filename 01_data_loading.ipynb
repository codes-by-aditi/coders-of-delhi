## 🎯 What is CodeBook?

CodeBook is a fictional social network for programmers in Delhi.  
Think of it like Facebook — but only for coders.

Each user on CodeBook has:
- A **profile** (name, city, interests)
- A **friends list** (people they're connected to)
- **Pages they follow** (tech communities, coding groups)

---

## 🎯 Project Goal

In this project, we will build **two smart features** for CodeBook:

| Feature | Logic |
|---|---|
| **People You May Know** | Suggest users based on mutual friends |
| **Pages You Might Like** | Suggest pages based on what friends like |

Both of these are real algorithms used by LinkedIn, Facebook, and Instagram.
