## 🎯 What is 'People You May Know'?

This is the friend suggestion feature you see on LinkedIn, Facebook, and Instagram.

The idea is simple:
> **If you and another person share many common friends, you probably know each other in real life.**

---

## 🧠 Algorithm: Mutual Friends

```
For every user in the network:
  1. Get the target user's friend list → Set A
  2. Get the candidate user's friend list → Set B
  3. Mutual Friends = A ∩ B  (set intersection)
  4. Score = len(Mutual Friends)
  5. Skip if candidate is already a friend of the target
  6. Rank all candidates by Score (highest first)
  7. Return top N
```

**Why set intersection?**  
Because sets in Python give us O(n) intersection — fast and clean.

## 🔍 Step 1 — Manual Example: Understanding Mutual Friends

Before running the full algorithm, let's understand it step by step  
with a concrete example.

**Question:** Do Aarav (u001) and Ananya (u007) share mutual friends?

````python
# Pick two users manually
aarav  = user_map.get('u001')
ananya = user_map.get('u007')

print(f'👤 {aarav["name"]} (u001) is friends with: {aarav["friends"]}')
print(f'👤 {ananya["name"]} (u007) is friends with: {ananya["friends"]}')

# Compute set intersection manually
set_a = set(aarav['friends'])
set_b = set(ananya['friends'])
mutual = set_a & set_b

print(f'\n  Set A (Aarav friends): {set_a}')
print(f'  Set B (Ananya friends): {set_b}')
print(f'  A ∩ B (Mutual friends): {mutual}')

if mutual:
    names = [user_map[fid]['name'] for fid in mutual if fid in user_map]
    print(f'\n  ✅ They share {len(mutual)} mutual friend(s): {names}')
    print(f'  → Aarav should be suggested to Ananya (and vice versa)!')
else:
    print('  ❌ No mutual friends found.')
````
